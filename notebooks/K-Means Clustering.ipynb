```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('data/Mall_Customers.csv')
```

```python
data.head()
```

```python
X=data.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i,init='k-means++')
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

```python
km=KMeans(n_clusters=5,init='k-means++',random_state=0)
y_km=km.fit_predict(X)
```

```python
plt.scatter(X[y_km==0,0],X[y_km==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(X[y_km==1,0],X[y_km==1,1],s=100,c='black',label='Cluster 2')
plt.scatter(X[y_km==2,0],X[y_km==2,1],s=100,c='blue',label='Cluster 3')
plt.scatter(X[y_km==3,0],X[y_km==3,1],s=100,c='yellow',label='Cluster 4')
plt.scatter(X[y_km==4,0],X[y_km==4,1],s=100,c='orange',label='Cluster 5')
plt.legend()
```
