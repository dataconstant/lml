```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('data/Mall_Customers.csv')
```

```python
X=data.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(X,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
ac=AgglomerativeClustering(n_clusters=5,linkage='ward',affinity='euclidean')
y_ac=ac.fit_predict(X)
```

```python
plt.scatter(X[y_ac==0,0],X[y_ac==0,1],c='red')
plt.scatter(X[y_ac==1,0],X[y_ac==1,1],c='blue')
plt.scatter(X[y_ac==2,0],X[y_ac==2,1],c='orange')
plt.scatter(X[y_ac==3,0],X[y_ac==3,1],c='green')
plt.scatter(X[y_ac==4,0],X[y_ac==4,1],c='black')
```
