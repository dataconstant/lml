```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

```python
train=pd.read_csv('train.csv')
```

```python
train.head()
```

## Explanatory Data Analysis

```python
train.count()
```

##### Checking missing data

```python
sns.heatmap(data=train.isnull(),yticklabels=False)
```

##### Survival count based on Class

```python
sns.countplot(x='Pclass',data=train,hue='Survived')
```

##### Survival count based on Sex 

```python
sns.countplot(x='Sex',data=train,hue='Survived')
```

###### Age Distribution

```python
sns.distplot(train['Age'].dropna(),kde=False)
```

##### Effect of sibling/spouse on Survival

```python
sns.countplot(x='SibSp',data=train,hue='Survived')
```

##### Effect of parents/siblings on Survival

```python
sns.countplot(x='Parch',data=train,hue='Survived')
```

##### Mean Age in each Class and Sex

```python
sns.boxplot(x='Pclass',y='Age',data=train,hue='Sex')
```
